## DATASET

### Descripcion

1 - Descripcion de datos
- Informacion de reservas de habitaciones de clientes de un hotel.  
<br><br>


2 - Diccionario de datos
- Atributos – Descripción
    - Booking_ID: identificador unico de registro
    - no_of_adults: Numero de adultos en la reserva
    - no_of_children: Numero de niños en la reserva
    - no_of_weekend_nights: Número de noches de fin de semana (sábado o domingo) que el huésped se alojó o reservó para alojarse en el hotel
    - no_of_week_nights: Número de noches entre semana (de lunes a viernes) que el huésped se alojó o reservó para alojarse en el hotel
    - type_of_meal_plan: Tipo de plan de comidas reservado por el cliente
    - required_car_parking_space: ¿Necesita el cliente una plaza de aparcamiento? (0 - No, 1 - Sí)
    - room_type_reserved: Tipo de habitación reservada por el cliente.
    - lead_time: Número de días entre la fecha de reserva y la fecha de llegada
    - arrival_year: Año de fecha de llegada
    - arrival_month: Mes de fecha de llegada
    - arrival_date: fecha del mes
    - market_segment_type: Designación del segmento de mercado.
    - repeated_guest: ¿Es el cliente un huésped recurrente? (0 - No, 1 - Sí)
    - no_of_previous_cancellations: Número de reservas anteriores que fueron canceladas por el cliente antes de la reserva actual
    - no_of_previous_bookings_not_canceled: Número de reservas anteriores no canceladas por el cliente antes de la reserva actual
    - avg_price_per_room: Precio medio por día de reserva; Los precios de las habitaciones son dinámicos. (en euros)
    - no_of_special_requests: Número total de solicitudes especiales realizadas por el cliente (por ejemplo, piso alto, vista desde la habitación, etc.)
    - booking_status: Bandera que indica si la reserva fue cancelada o no.

### Lectura e informacion

```python
#IMPORTACION DE LIBRERIA
import pandas as pd

#LECTURA DE DATASET
df = pd.read_csv("https://raw.githubusercontent.com/arielRas/DataSets/main/Hotel%20Reservations.csv")
df.head(3)
```

```python
#SE IMPRIME INFORMACION DEL DATASET
print(f"Filas y columnas: {df.shape}")

print("---------------------------------------------------------------")
print(df.info())
```

```python
#MAS INFORMACION
df.describe().round(2)
```

### Limpieza y transformacion

```python
#SE TRANSFORMA LOS VALORES DE LA VARIABLE DE INTERES A NUMERICO

print(f"Valores en la variable de interes: {df.booking_status.unique()}")

df["booking_status"] = df["booking_status"].replace('Not_Canceled',0)
df["booking_status"] = df["booking_status"].replace("Canceled", 1)
df["booking_status"] = df["booking_status"].astype(int)

print(f"Valores en la variable de interes: {df.booking_status.unique()}")
```

```python
#SE AGREGA COLUMNA CON FECHA COMPLETA 

df['arriv_date_compl'] = df['arrival_year'].astype(str) + "/" + df['arrival_month'].astype(str) + "/" + df['arrival_date'].astype(str)

df['arriv_date_compl'] = pd.to_datetime(df['arriv_date_compl'], format="%Y/%m/%d", yearfirst=True, errors='coerce')

df[['arrival_year','arrival_month','arrival_date','arriv_date_compl']].head(3)
```

```python
#SE VERIFICA SI SE GENERARON DATOS NAT INCOMPATIBLES CON DATETIME

print(f"Cantidad de datos Nat que se generaron: {df.arriv_date_compl.isnull().sum()}")
```

```python
#SE ELIMINAN LOS DATOS NAT QUE SE GENERARON
df.dropna(subset=['arriv_date_compl'], inplace=True)

#SE RECONVIERTE EN DATETIME LA COLUMNA
df['arriv_date_compl'] = pd.to_datetime(df['arriv_date_compl'])

#VERIFICACION DE DATOS NAT Y DE TAMAÑO DE DATASET LUEGO DE ELIMINAR
print(f"Cantidad de datos Nat: {df.arriv_date_compl.isnull().sum()}")
print(f"Cantidad de filas y columnas luego de limpiar: {df.shape}")
df[['arrival_year','arrival_month','arrival_date','arriv_date_compl']].head(3)
```

El objetivo de crear una columna de fecha completa, es para poder establecer la estacion del año a la que pertenece cada fecha

```python
#CREANDO ESTACIONES DEL AÑO (EUROPA)

#Creacion de variables
month = df['arriv_date_compl'].dt.month
date = df['arriv_date_compl'].dt.day

#se crean variables bcon condiciones a fin de establecer estaciones
is_summer = ((month == 6) & (date >= 21) | (month == 7) | (month == 8) | (month == 9) & (date <= 22))
is_autumn = ((month == 9) & (date >= 23) | (month == 10) | (month == 11) | (month == 12) & (date <= 20))
is_winter = ((month == 12) & (date >= 21) | (month == 1) | (month == 2) | (month == 3) & (date <= 20))

#Se crea la columna "Season" y se le asigna el valor "spring" para todos los valores de la columna
df["season"] = "spring"

#Se asigna valores a la columna "season" segun la fecha
df.loc[is_summer, "season"] = "summer"
df.loc[is_autumn, "season"] = "autumn"
df.loc[is_winter, "season"] = "winter"

df[['arriv_date_compl',"season"]].head(3)
```

## ML

### Importacion de librerias

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

### Tratamiento de variables categoricas

```python
#VALORES DE LAS VARIABLES CATEGORICAS
print(df.type_of_meal_plan.unique())
print(df.room_type_reserved.unique())
print(df.market_segment_type.unique())
print(df.season.unique())
```

Para este caso especifico se utiliza label encoder con la finalidad de otorgarle un valor numerico a las variables categoricas. La eleccion de esta herramienta se debe a que los valores unicos que existen para las variables son relativamente pocos.

```python
encoder = LabelEncoder()

df.insert(loc=df.columns.get_loc("type_of_meal_plan")+ 1, column="meal_plan_cod", value=encoder.fit_transform(df.type_of_meal_plan.values))

df.insert(loc=df.columns.get_loc("room_type_reserved")+ 1, column="room_type_cod", value=encoder.fit_transform(df.room_type_reserved.values))

df.insert(loc=df.columns.get_loc("market_segment_type")+ 1 , column="market_segment_cod", value=encoder.fit_transform(df.market_segment_type.values))

df.insert(loc=df.columns.get_loc("season")+ 1, column="season_cod", value=encoder.fit_transform(df.season.values))

df.head(5)
```

### Preparacion de datos y ejecucion del algoritmo

```python
X = df.drop(["Booking_ID", "type_of_meal_plan", "room_type_reserved", "market_segment_type", "season", "arrival_year", "arrival_month","arriv_date_compl","booking_status"], axis=1)
y = df["booking_status"]
```

```python

# Ajustar un árbol de decisión con hiperparámetros por defecto
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=1234)
clf = DecisionTreeClassifier(random_state=1234, max_depth=3)
model = clf.fit(X_train, y_train)
predict = clf.predict(X_test)

columns_name = X.columns.to_list()
class_name = y.astype(str).unique().tolist()

# Graficar el árbol de decisión
fig = plt.figure(figsize=(15, 7))
_ = tree.plot_tree(clf, feature_names=columns_name, class_names=class_name, filled=True)
plt.show()
```

### Metricas de Clasificacion

```python
print(f"Precicion del modelo en entrenamiento: {model.score(X_train,y_train)}")
print(f"Precicion del modelo en etapa de testing: {model.score(X_test, y_test)}")
print(f"Precicion del modelo en etapa de testing: {metrics.accuracy_score(y_test,model.predict(X_test))}")
print("--------------------------------------------------------\n")

print(metrics.classification_report(y_test, predict))
```
